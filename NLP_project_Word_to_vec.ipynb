### Remove Punctuation

```python
import string
string.punctuation
```

```python
def remove_punct(text):
  text = "".join([char for char in text if char not in string.punctuation])
  return text

messages['text_clean'] = messages['text'].apply(lambda x: remove_punct(x))
messages.head()
```

### Tokenization

```python
import re

def tokenize(text):
  tokens = re.split('\W+', text)
  return tokens

messages['text_tokenized'] = messages['text_clean'].apply(lambda x: tokenize(x.lower()))

messages.head()
```

### Remove Stopwrods

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
```

```python
# Define a function to remove all stopwords
def remove_stopwords(tokenized_text):
  text = [word for word in tokenized_text if word not in stopwords]
  return text

messages['text_nostop'] = messages['text_tokenized'].apply(lambda x: remove_stopwords(x))

messages.head()
```

```python
# Remove stopwords in the example
remove_stopwords(tokenize("I am learning NLP".lower()))
```

```python
# Define a function to handle all data cleaning
def clean_text(text):
  text = "".join([word.lower() for word in text if word not in string.punctuation])
  tokens = re.split("\W+", text)
  text = [word for word in tokens if word not in stopwords]
  return text
```

```python
# Fit  a basic TFIDF Vectorizer and view the results
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(messages['text'])
print(X_tfidf.shape)
print(tfidf_vect.get_feature_names_out())
```

```python
X_tfidf
```

```python
X_features = pd.DataFrame(X_tfidf.toarray())
X_features.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
print(RandomForestClassifier())
```

### Explore RandomForestClassifier On a Handout Set

```python
from sklearn.metrics import precision_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
# Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_features,
                                                    messages['label'],
                                                    test_size = 0.2)
```

```python
# Fit a basic Random Forest Model
rf = RandomForestClassifier()
rf_model = rf.fit(X_train, y_train)
```

```python
# Make predictions on the test set using the fit model
y_pred = rf_model.predict(X_test)
```

```python
# Evaluate model predictions using prediction and recall
precision = precision_score(y_test, y_pred, pos_label = 'spam')
recall = recall_score(y_test, y_pred, pos_label = 'spam')
print('Precision: {} / Recall: {}'.format(round(precision, 3), round(recall, 3)))
```

### Train the model

```python
import gensim
messages['text_clean'] = messages['text'].apply(lambda x: gensim.utils.simple_preprocess(x))
messages.head()
```

```python
messages.drop(['text_tokenized', 'text_nostop'], axis=1, inplace=True)
messages.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(messages['text_clean'],
                                                    messages['label'], test_size = 0.2)
```

```python
from gensim.models import Word2Vec
# Assuming X_train is a list of tokenized sentences
w2v_model = Word2Vec(sentences=X_train,
                     vector_size=100,
                     window=5,
                     min_count=2)
```

```python
#Explore the word vector for 'king' based on the trained model
w2v_model.wv['king']
```

```python
# Find the most similar words to 'King' based on word vectors from the trained model
w2v_model.wv.most_similar('king')
```

```python
w2v_model.wv.index_to_key
```

```python
import numpy as np
```

```python
w2v_vect = []
for ls in X_test:
    vectors = [w2v_model.wv[i] for i in ls if i in w2v_model.wv.key_to_index]
    if vectors:
        w2v_vect.append(np.array(vectors))
    else:
        # If no word vectors are found for the sentence,append a zero vector
        w2v_vect.append(np.zeros((1, w2v_model.vector_size)))

# Filter out empty arrays (those with zero vectors)
w2v_vect = [seq for seq in w2v_vect if seq.size > 0]

# Convert the list to a numpy array
w2v_vect = np.array(w2v_vect, dtype=object)

print(w2v_vect)
```

```python
# Why is the length of the sentence different than the length of the sentence vector?
for i,v in enumerate(w2v_vect):
    print(len(X_test.iloc[i]), len(v))
```

```python
# Compute sentence vectors by averaging the word vectors for the words contained in the sentence
w2v_vect_avg = []

for vect in w2v_vect:
    if len(vect)!=0:
        w2v_vect_avg.append(vect.mean(axis=0))
    else:
        w2v_vect_avg.append(np.zeros(100))
```

```python
# Are our sentence vector lengths consistent?
for i,v in enumerate(w2v_vect_avg):
    print(len(X_test.iloc[i]), len(v))
```
